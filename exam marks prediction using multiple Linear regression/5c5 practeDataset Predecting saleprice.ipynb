```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#reading data set
dataset=pd.read_csv("practiceDataset.csv")
dataset
```

```python
dataset.shape #it contains 81 columns and 1460 rows
```

```python
dataset.info()
```

```python
dataset.columns[dataset.isna().any()]  #this all columns contains "Nan" values that means missing values
```

```python
#let us take few columnsand predict the saleprice
df=dataset[['MSSubClass', 'MSZoning', 'LotFrontage', 'LotArea', 'HouseStyle', 'OverallQual', 'OverallCond','SalePrice']]
df
```

```python
#from the above dataset we need to predict the saleprice using indeependent variable
#finding missing values columns
df.columns[df.isna().any()]
```

```python
#we can observe that LotFrontage contains missing value so that we neeed to handle it by replace with mean of that colums
df.LotFrontage=df.LotFrontage.fillna(df.LotFrontage.mean())
df
```

```python
df.isna().count() #there is no missing values in the df dataset
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.MSZoning=le.fit_transform(df.MSZoning)
df.HouseStyle=le.fit_transform(df.HouseStyle)
```

```python
#independent and dependent variables
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```
